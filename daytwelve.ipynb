```python
from functools import cache


input = open('inputs/d7input.txt', 'r')
lines = input.readlines()

@cache
def checkPlan(cogs:str) -> tuple:
    coggers = [i for i in filter(lambda x: x != '', cogs.split('.'))]
    return tuple([len(i) for i in coggers])

@cache
def recurse(front: str, plan: tuple, cogs : str) -> int:
    if len(cogs) == 0:
        if len(plan) == 0:
            print("yes to cogs") 
        return 1 if len(plan) == 0 else 0
    if '?' not in cogs:
        return 1 if checkPlan(cogs) == plan else 0
    else:
        index = cogs.index('?')
        left = cogs[:index]
        run = 0
        for i in range(len(left)):
            if left[i] == '.':
                if run > 0:
                    if run == plan[0]:
                        plan = plan[1:]
                        if(len(plan) == 0):
                            return 1 if '#' not in cogs[i:] else 0
                        run = 0
                    else:
                        return 0
            else:
                run += 1
        right = cogs[index+1:]
        if(run > 0):
            if run == plan[0]:
                if len(plan) == 1:
                    return 1 if '#' not in right else 0
                return recurse(left, plan[1:], right)
            elif run < plan[0]:
                return recurse(left[:-(run)], plan, '#'*(run+1) + right)
            else:
                return 0
        else:
            return recurse(left[:-1], plan, '#' + right) + recurse(left, plan, right)

cum_sum = 0
for i in range(len(lines)):
    ln = lines[i].strip()
    split = ln.split(' ')
    cogs = '?'.join([split[0] for _ in range(5)])
    preplan = [int(i) for i in split[1].split(',')]
    plan = tuple(preplan*5)
    print(recurse('',plan,cogs))
    cum_sum += recurse('',plan,cogs)

cum_sum
```
