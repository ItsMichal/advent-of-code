```python
import functools


input = open('inputs/d7input.txt', 'r')
lines = input.readlines()

order = [' ', 'F', 'R', 'H', 'E', 'D', 'U', 'N', 'A', 'K', 'Q', 'T', '9', '8', '7', '6', '5', '4', '3', '2', 'J']

ranking = []

def sortBySuit(a , b):
    # find a in order
    aIndex = order.index(a)
    bIndex = order.index(b)

    return aIndex - bIndex

def sortByHand(a,b):
    index = 0
    while(a[0][index] == b[0][index]):
        if index == len(a[0])-1:
            return 0
        index += 1
    return sortBySuit(a[0][index], b[0][index])

def sortHand(hand):
    # sort by rank
    hand = sorted(hand, key=functools.cmp_to_key(sortBySuit))
    collapsed = {}
    for card in hand:
        if card in collapsed:
            collapsed[card] += 1
        else:
            collapsed[card] = 1
    
    # remove jokers
    if 'J' in collapsed:
        # remove joker if not all 5 jokes (what an edge case)
        if collapsed['J'] != 5:
            # add jokers to highest card count
            joker_count = collapsed['J']
            del collapsed['J']

            # find highest card count
            highest = 0
            hkey = ''
            for key in collapsed:
                if collapsed[key] > highest:
                    highest = collapsed[key]
                    hkey = key
            
            # add jokers to highest card count
            collapsed[hkey] += joker_count

    # five of a kind
    if 5 in collapsed.values():
        return 'F '
    # four of a kind
    elif 4 in collapsed.values():
        return 'R '
    # full house
    elif 3 in collapsed.values() and 2 in collapsed.values():
        return 'H '
    # three of a kind
    elif 3 in collapsed.values():
        return 'E '
    # two pair
    elif list(collapsed.values()).count(2) == 2:
        return 'D '
    # one pair
    elif 2 in collapsed.values():
        return 'U '
    
    return 'N '
    
    # high card
    return hand[0]+' '

for i in range(len(lines)):
    ln = lines[i].replace('\n', '')
    split = ln.split(' ')
    
    bid = int(split[1])
    hand = split[0].strip()
    
    hand = sortHand(hand)+hand
    ranking.append((hand, bid))

ranking.sort(key=functools.cmp_to_key(sortByHand))

cum = 0
for i in range(len(ranking)):
    reward = i+1
    cum += reward * ranking[len(ranking)-i-1][1]

print(cum)

```
