```python
input = open('inputs/d2input.txt', 'r').readlines()

def validate_game(str):
    # split on comma
    split = str.split(',')
    for i in range(len(split)):

        #split on space
        splitted = split[i].strip().split(' ')

        # print(splitted)

        if(splitted[1] == 'green'):
            if(int(splitted[0]) > 13):
                return False
        elif(splitted[1] == 'blue'):
            if(int(splitted[0]) > 14):
                return False
        elif(splitted[1] == 'red'):
            if(int(splitted[0]) > 12):
                return False
    
    return True

def get_min_from_game(str):
    split = str.split(',')
    minr = 0
    ming = 0
    minb = 0

    for i in range(len(split)):

        #split on space
        splitted = split[i].strip().split(' ')

        # print(splitted)

        if(splitted[1] == 'green'):
            if(int(splitted[0]) > ming):
                ming = int(splitted[0])
        elif(splitted[1] == 'blue'):
            if(int(splitted[0]) > minb):
                minb = int(splitted[0])
        elif(splitted[1] == 'red'):
            if(int(splitted[0]) > minr):
                minr = int(splitted[0])

    return (minr, ming, minb)

      
cum_sum = 0
for i in range(len(input)):
    line = input[i]

    # split on colon
    split = line.split(':')

    #get id of line by splitting game
    id = split[0].split(' ')[1]

    # get list of games by splitting semicolons

    game_list = split[1].split(';')

    valid = True
    for j in range(len(game_list)):
        game = game_list[j]
        if(not validate_game(game)):
            valid = False
            break

    if(valid):
        cum_sum+=int(id)
        print(id)

print(cum_sum)
```
