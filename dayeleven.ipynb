```python
input = open('inputs/d7input.txt', 'r')
lines = input.readlines()

columns = []
rows = []

new_data = []
parsing = True

row_index = 0

galaxies = []

while parsing:
    ln = lines[row_index].strip()
    if row_index == 0:
        columns = [True for i in range(len(ln))]
    
    if '#' in ln:
        rows.append(False)
    else:
        rows.append(True)
    
    # check cols
    for i in range(len(ln)):
        if ln[i] == '#':
            columns[i] = False
    
    new_data.append(ln)
    print(ln)
    
    if row_index == len(lines) - 1:
        parsing = False
    else:
        row_index += 1

for i in range(len(new_data)):
    print(new_data[i])
    for j in range(len(new_data[i])):
        if new_data[i][j] == '#':
            galaxies.append((i, j))
    
def getNewIdx(galaxy):
    mult = 1000000
    a,b = (0,0)
    for i in range(galaxy[0]):
        if rows[i]:
            a+=mult
        else:
            a+=1
    for i in range(galaxy[1]):
        if columns[i]:
            b+=mult
        else:
            b+=1
    return (a,b)

def manhattan_distance(a, b):
    new_a = getNewIdx(a)
    new_b = getNewIdx(b)
    return abs(new_a[0] - new_b[0]) + abs(new_a[1] - new_b[1])
    

cum_sum = 0
for i in range(len(galaxies)):
    for j in range(len(galaxies)):
        if i != j:
            cum_sum += manhattan_distance(galaxies[i], galaxies[j])

cum_sum/2
```
