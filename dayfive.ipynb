The almanac starts by listing which seeds need to be planted: seeds 79, 14, 55, and 13.

The rest of the almanac contains a list of maps which describe how to convert numbers from a source category into numbers in a destination category. That is, the section that starts with seed-to-soil map: describes how to convert a seed number (the source) to a soil number (the destination). This lets the gardener and his team know which soil to use with which seeds, which water to use with which fertilizer, and so on.

Rather than list every source number and its corresponding destination number one by one, the maps describe entire ranges of numbers that can be converted. Each line within a map contains three numbers: the destination range start, the source range start, and the range length.

Consider again the example seed-to-soil map:

50 98 2
52 50 48
The first line has a destination range start of 50, a source range start of 98, and a range length of 2. This line means that the source range starts at 98 and contains two values: 98 and 99. The destination range is the same length, but it starts at 50, so its two values are 50 and 51. With this information, you know that seed number 98 corresponds to soil number 50 and that seed number 99 corresponds to soil number 51.

The second line means that the source range starts at 50 and contains 48 values: 50, 51, ..., 96, 97. This corresponds to a destination range starting at 52 and also containing 48 values: 52, 53, ..., 98, 99. So, seed number 53 corresponds to soil number 55.

Any source numbers that aren't mapped correspond to the same destination number. So, seed number 10 corresponds to soil number 10.

So, the entire list of seed numbers and their corresponding soil numbers looks like this:
```
seed  soil
0     0
1     1
...   ...
48    48
49    49
50    52
51    53
...   ...
96    98
97    99
98    50
99    51
```
With this map, you can look up the soil number required for each initial seed number:

- Seed number 79 corresponds to soil number 81.
- Seed number 14 corresponds to soil number 14.
- Seed number 55 corresponds to soil number 57.
- Seed number 13 corresponds to soil number 13.

## --- Part Two ---
Everyone will starve if you only plant such a small number of seeds. Re-reading the almanac, it looks like the seeds: line actually describes ranges of seed numbers.

The values on the initial seeds: line come in pairs. Within each pair, the first value is the start of the range and the second value is the length of the range. So, in the first line of the example above:
```
seeds: 79 14 55 13
```
This line describes two ranges of seed numbers to be planted in the garden. The first range starts with seed number 79 and contains 14 values: 79, 80, ..., 91, 92. The second range starts with seed number 55 and contains 13 values: 55, 56, ..., 66, 67.

Now, rather than considering four seed numbers, you need to consider a total of 27 seed numbers.

In the above example, the lowest location number can be obtained from seed number 82, which corresponds to soil 84, fertilizer 84, water 84, light 77, temperature 45, humidity 46, and location 46. So, the lowest location number is 46.

Consider all of the initial seed numbers listed in the ranges on the first line of the almanac. What is the lowest location number that corresponds to any of the initial seed numbers?


````python
from aocd import data, submit

lines = data.split('\n')
total = 0

# get seeds
seeds = lines[0].split(":")[1].split(' ')[1:]
seeds = [int(i) for i in seeds]

print(seeds)

ranges = []

for i in range(0, len(seeds), 2):
    ranges.append([seeds[i], seeds[i]+seeds[i+1]-1])

isOnMapLine = False
converted = []
print(ranges)

for i in range(len(lines)):
    # strip any annoying chars
    ln = lines[i].strip()

    # if we are in a map line
    if isOnMapLine:
        # check to see if we ran out of map lines
        if ln != '':
            
            # get the destination, src, and maximum offset from the map
            dest = int(ln.split(' ')[0])
            src = int(ln.split(' ')[1])
            max_off = int(ln.split(' ')[2])

            # set up a queue to run through each number range and update it with the new possible ranges
            j=0
            while(j < len(ranges)):

                # get minimum and maximum of these ranges
                # while the problem excludes the max, 
                # I like thinking inclusively, so 
                # my ranges are inclusive
                min = ranges[j][0]
                max = ranges[j][1]


                # here we will filter into 6 cases
                # this can easily be skipped/optimized
                # but this is way more readable/clean/understandable

                # ideally, we never keep this default case
                # we should throw an error if we see it-
                # but not in leetcode
                case = "Unknown"

                # Check if the entire mapped range is included
                # within our min/max. This will result in 3 new ranges
                # one converted range, and two (lower, upper) untouched ranges
                if(src > min and src+max_off-1 < max):
                    case = "BothInside"
                
                # Similarly, if our range is the same as the mapped range
                # or encapsulated by it, we should convert the whole range
                elif(src <= min and src+max_off-1 >= max):
                    case = "Encapsulated"
                
                # If the mapped range's upper bound is inside, at this point
                # the lower bound must be outside (otherwise, we'd hit the BothInside check)
                # so, we convert everything below the upper bound and 
                # leave the top half untouched
                elif(src+max_off-1 > min and src+max_off-1 < max):
                    case = "UpperInside"
                # very similar logic to UpperInside, but flipped so we
                # convert the top half and leave the bottom half
                elif(src > min and src < max):
                    case = "LowerInside"
                
                # At this point, we can assume that all other cases
                # do not contain our range, and are outside. Here
                # are two checks to just classify on which side they are.
                elif(src < min and src+max_off-1 < min):
                    case = "OutsideLower"
                elif(src > max and src+max_off-1 > max):
                    case = "OutsideHigher"
                

                # We implement the logic discussed above here.
                # We could couple this logic with the if check
                # But this helped me make sure I made no errors
                # in my logic.

                # We use two arrays (ranges, converted) as once
                # a range has matched a rule, it should not be 
                # compared again within the same "map block".

                # We store mapped ranges in converted, and
                # anything untouched in ranges.

                # One important realization is that you should
                # "recycle" unmatched ranges from this step back
                # into "ranges" -> hence why we use a queue-like
                # instead of just a simple itteration. We can re-add
                # new chunks of ranges and see if they match future rules

                # Not realizing the above cost me a good bit of time in AoC!
                if(case == "BothInside"):
                    # Pop the original range, add the top/bot half back in,
                    # add add the converted middle chunk
                    ranges.pop(j)
                    ranges.extend(
                        [
                            [min, src],
                            [src+max_off, max]
                        ]
                    )
                    converted.extend(
                        [
                            [dest, dest+max_off-1],
                        ]
                    )
                elif(case == "UpperInside"):
                    # Similar, but we have two halves now, the unmapped above the upper limit
                    # and the mapped below it
                    ranges.pop(j)
                    ranges.extend([
                        [src+max_off, max]
                    ])
                    converted.extend([
                        [min+(dest-src), dest+max_off-1],
                    ])
                elif(case == "LowerInside"):
                    # Same but other way around
                    ranges.pop(j)
                    ranges.extend([
                        [min, src-1]
                    ])
                    converted.extend([
                        [dest, max+(dest-src)]
                    ])
                elif(case == "Encapsulated"):
                    # Convert everything!
                    ranges.pop(j)
                    converted.extend([
                        [min+(dest-src), max+(dest-src)]
                    ])
                else:
                    # If we are outside, we increment the index
                    # Convert nothing.
                    j+=1

        # if we are done with mapping for this block
        # aka lines == '' 
        else:
            # extend the converted with whatever we didn't manage to convert.
            # now that I think of it, reversing this saves a line of code...
            converted.extend(ranges)
            ranges = converted

            # reset the converted for the next block
            converted=[]

            # tell the parser to look for lines with "map" again
            isOnMapLine=False

    
    #finally check if this line indicates a map
    #if so, we will treat the next rows as containing
    #mapping data
    if "map" in ln:
      isOnMapLine = True
      pass

# Finally, we do this unoptimal combination again
converted.extend(ranges)
ranges = converted

# and we do a custom sort on the min of every range
ranges.sort(key=lambda x: x[0])

# and submit the smallest minimum we have- which is the answer to the minimum distance!
submit(ranges[0][0])
````
